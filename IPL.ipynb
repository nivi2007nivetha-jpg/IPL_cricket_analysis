```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("df_batting.csv")

data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
player_runs = data.groupby("batsmanName")["runs"].sum().sort_values(ascending=False)

player_runs.head(10)
```

```python
top5 = player_runs.head(5)

plt.figure()
top5.plot(kind="bar")
plt.title("Top 5 Players by Total Runs")
plt.xlabel("Player")
plt.ylabel("Total Runs")
plt.show()
```
